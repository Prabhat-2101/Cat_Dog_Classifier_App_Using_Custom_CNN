```python
import matplotlib.pyplot as plt
from PIL import Image
import random
import os
import seaborn as sns
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array
from keras.utils import plot_model
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
```

```python
cat_directory_path = '/content/dogs_vs_cats/train/cats'
dog_directory_path = '/content/dogs_vs_cats/train/dogs'
```

```python
cat_image_files = os.listdir(cat_directory_path)
dog_image_files = os.listdir(dog_directory_path)

print(len(cat_image_files),len(cat_image_files))
```

**Checking for Invalid input file extension.**

```python
allowed_file = ['.jpeg', '.jpg', '.png']
cat_count = sum(1 for file_name in cat_image_files if os.path.splitext(file_name)[1] in allowed_file)
dog_count = sum(1 for file_name in dog_image_files if os.path.splitext(file_name)[1] in allowed_file)
print(list([cat_count,dog_count]))
```

**Create Imagedatagenerator and pass images in batches.**

```python
base_dir = "/content/dogs_vs_cats/train"
test_dir = '/content/dogs_vs_cats/test'
training_datagenerator = ImageDataGenerator(rescale=1./255,validation_split=0.2)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = training_datagenerator.flow_from_directory(
    base_dir,
    target_size=(150,150),
    batch_size=20,
    class_mode="binary",
    shuffle=True,
    subset="training",
)
valid_generator = training_datagenerator.flow_from_directory(
    base_dir,
    target_size=(150,150),
    batch_size=20,
    class_mode="binary",
    shuffle=True,
    subset="validation",
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=20,
    class_mode='binary',
    shuffle=False
)
```

```python
train_labels = train_generator.class_indices
valid_labels = valid_generator.class_indices
print("Training class labels:", train_labels)
print("Validation class labels:", valid_labels)
```

**Create a Custom CNN model**

```python
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.summary()
```

```python
plot_model(model,to_file="model_architecture.png",show_shapes=True,show_layer_names=True)
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
batch_size = 20
train_steps_per_epoch = np.ceil(train_generator.samples / batch_size)
validation_steps_per_epoch = np.ceil(valid_generator.samples / batch_size)
test_steps=np.ceil(test_generator.samples/test_generator.batch_size)
print(train_steps_per_epoch, validation_steps_per_epoch,test_steps)
```

```python
history = model.fit(
              train_generator,
              steps_per_epoch=800,
              epochs=5,
              validation_data=valid_generator,
              validation_steps=200
          )
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(len(acc))

plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
loss, accuracy = model.evaluate(test_generator)
print(f"Test accuracy: {accuracy*100:.2f}%")
print(f"Test loss: {loss:.2f}")
```

```python
probabilities = model.predict(test_generator,steps=test_steps)
predicted_classes = (probabilities > 0.5).astype(int).squeeze()
class_labels = list(test_generator.class_indices.keys())
predicted_classes
```

```python
cm = confusion_matrix(test_generator.classes, predicted_classes)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

**Predicting Some Sample Images of Cats and Dogs**

---



```python
def make_prediction(image_path):
    img = load_img(image_path, target_size=(150, 150))
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.

    prediction = model.predict(img_array)
    print("Prediction:", prediction)

    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    if prediction[0][0] > 0.5:
        plt.title('Predicted: Dog')
    else:
        plt.title('Predicted: Cat')
    plt.imshow(img)  # Show the same image as prediction
    plt.axis('off')
    plt.show()
```

```python
img_path = '/content/cat.jpg'
make_prediction(img_path)
```
